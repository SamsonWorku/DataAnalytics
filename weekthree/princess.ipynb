```python
princess_wkt = []
with open("datasets_12dancingprincesses.txt", 'r') as file:
    for line in file:
        cline = line.strip() #get rid of newline character

        if cline == '': pass  #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline) #tokenize each part of the the text
            #print(tknls)
        for word in tknls: #token list
            princess_wkt.append(word)

len(princess_wkt)
#princess_wkt
#tknls
#print(tknls)
```

```python
princess_wkt
```

```python
#can't run this before removing additional words UNLESS save it into another var
princess_freq = FreqDist(princess_wkt)
princess_freq
```

```python
#p_no_punc = []
princess_no_punc = princess_wkt
for word in princess_no_punc:
    if word in punctuation: #punctuation is a imported function
        princess_no_punc.remove(word) #no stop words
```

```python
#princess_no_sw = princess_no_punc
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python
rm_count = 0
princess_no_sw = []

for word in princess_no_punc:
    if word not in eng_stopwords:
        princess_no_sw.append(word)
    else:
        rm_count += 1
```
